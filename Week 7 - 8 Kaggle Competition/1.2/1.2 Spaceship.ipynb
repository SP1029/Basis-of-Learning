```python
def f(x):
    x = np.floor(x+0.5)
    return x
```

```python
df = pd.read_csv("Final Data Train.csv")
df
```

```python
X = df.loc[0:8000,["PID","PID3/4","HP1","HP2","HP3","CryoSleep","DB","DC","DC4/5","Side","D1","D2","D3","NormAge","VIP","NormRoomService","NormFoodCourt","NormShoppingMall","NormSpa","NromVRDeck"]].to_numpy()
Y = df.loc[0:8000,["Transported"]].to_numpy()
X1 = df.loc[8001:8692,["PID","PID3/4","HP1","HP2","HP3","CryoSleep","DB","DC","DC4/5","Side","D1","D2","D3","NormAge","VIP","NormRoomService","NormFoodCourt","NormShoppingMall","NormSpa","NromVRDeck"]].to_numpy()
Y1 = df.loc[8001:8692,["Transported"]].to_numpy()
```

```python
def sigmoid (no):
    e = 2.718281828459045
    return 1/(1+e**(-no))
```

```python
def activ (no):
    e = 2.718281828459045
    return 1/(1+e**(-no))
```

```python
def derv (no):
    return no * ( 1 - no )
```

```python
def sigmoid_inv (no):
    return np.log(no/(1-no))
```

```python
N = 8000
M = 692
n0 = 20
n1 = 10
n2 = 8
n3 = 7
n4 = 6
n5 = 5
n6 = 1
w_0 = np.random.rand(n0,n1) - 0.5
w_1 = np.random.rand(n1,n2) - 0.5
w_2 = np.random.rand(n2,n3) - 0.5
w_3 = np.random.rand(n3,n4) - 0.5
w_4 = np.random.rand(n4,n5) - 0.5
w_5 = np.random.rand(n5,n6) - 0.5
b_0 = np.random.rand(1,n1) - 0.5
b_1 = np.random.rand(1,n2) - 0.5
b_2 = np.random.rand(1,n3) - 0.5
b_3 = np.random.rand(1,n4) - 0.5
b_4 = np.random.rand(1,n5) - 0.5
b_5 = np.random.rand(1,n6) - 0.5

# last column is coeff b/w ans and first layer
# In r th row, cell with column i is the weight from i th input to r th neuron in first layer
# Last row for bias
```

```python
alpha = 0.0001
```

```python
a = 0

while a < 100000:
      
    # data -> 1st Layer -> Final
    
    z_1 = np.dot( X , w_0 ) + b_0
    a_1 = activ( z_1 )
    z_2 = np.dot( a_1 , w_1 ) + b_1
    a_2 = activ( z_2 )
    z_3 = np.dot( a_2 , w_2 ) + b_2
    a_3 = activ( z_3 )
    z_4 = np.dot( a_3 , w_3 ) + b_3
    a_4 = activ( z_4 )
    z_5 = np.dot( a_4 , w_4 ) + b_4
    a_5 = activ( z_5 )
    z_6 = np.dot( a_5 , w_5 ) + b_5
    a_6 = activ( z_6 )
    
    # Cost Function
    
    
    Cost = np.sum((a_6 - Y) ** 2) * 1 / N
#     print(Cost)
    if a % 1 == 0:
#         print("\n")
#         print(np.sum(abs((f(a_6) - Y)*100))/N)
        answer1 = sigmoid(np.dot(X1,w_0) + b_0)
        answer2 = sigmoid(np.dot(answer1,w_1) + b_1)
        answer3 = sigmoid(np.dot(answer2,w_2) + b_2)
        answer4 = sigmoid(np.dot(answer3,w_3) + b_3)
        answer5 = sigmoid(np.dot(answer4,w_4) + b_4)
        answer6 = sigmoid(np.dot(answer5,w_5) + b_5)
        print (np.sum(abs((f(answer6) - Y1)))*100/M)
#         print("\n")

    # Safety
    
    if a == 0:
        cost_prev = Cost
    cost_curr = Cost
    if (cost_curr>cost_prev):
        alpha = alpha / 1.51
        cost_prev = cost_curr
        continue
    cost_prev = cost_curr
    
    # Optimise
    
    if a % 200 == 0:
        alpha = alpha * 1.5
    
    
    # Update
    
    da_6 = 2 * ( a_6 - Y )
    dz_6 = derv( a_6 ) * da_6
    dw_5 = np.dot( np.transpose(a_5) , dz_6 )
    db_5 = np.sum( dz_6 , keepdims = True)
    
    da_5 = np.dot( dz_6 , np.transpose(w_5) )
    dz_5 = derv( a_5 ) * da_5
    dw_4 = np.dot( np.transpose(a_4) , dz_5 )
    db_4 = np.sum( dz_5 , keepdims = True)
    
    da_4 = np.dot( dz_5 , np.transpose(w_4) )
    dz_4 = derv( a_4) * da_4
    dw_3 = np.dot( np.transpose(a_3) , dz_4 )
    db_3 = np.sum( dz_4 , keepdims = True)
    
    da_3 = np.dot( dz_4 , np.transpose(w_3) )
    dz_3 = derv( a_3 ) * da_3
    dw_2 = np.dot( np.transpose(a_2) , dz_3 )
    db_2 = np.sum( dz_3 , keepdims = True)
    
    da_2 = np.dot( dz_3 , np.transpose(w_2) )
    dz_2 = derv( a_2 ) * da_2
    dw_1 = np.dot( np.transpose(a_1) , dz_2 )
    db_1 = np.sum( dz_2 , keepdims = True)
    
    da_1 = np.dot( dz_2 , np.transpose(w_1) )
    dz_1 = derv( a_1 ) * da_1
    dw_0 = np.dot( np.transpose(X) , dz_1 )
    db_0 = np.sum( dz_1 , keepdims = True)
    
    w_0 = w_0 - alpha * dw_0
    w_1 = w_1 - alpha * dw_1
    w_2 = w_2 - alpha * dw_2
    w_3 = w_3 - alpha * dw_3
    w_4 = w_4 - alpha * dw_4
    w_5 = w_5 - alpha * dw_5
    
    b_0 = b_0 - alpha * db_0
    b_1 = b_1 - alpha * db_1
    b_2 = b_2 - alpha * db_2    
    b_3 = b_3 - alpha * db_3
    b_4 = b_4 - alpha * db_4
    b_5 = b_5 - alpha * db_5
    
    # Iteration
    
    a = a + 1
```
