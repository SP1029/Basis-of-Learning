```python
def f(x):
    x = np.floor(x+0.5)
    return x
```

```python
df = pd.read_csv("Final Data Train.csv")
df
```

```python
X = df.loc[0:8500,["PID","PID3/4","HP1","HP2","HP3","CryoSleep","DB","DC","DC4/5","Side","D1","D2","D3","NormAge","VIP","NormRoomService","NormFoodCourt","NormShoppingMall","NormSpa","NromVRDeck"]].to_numpy()
Y = df.loc[0:8500,["Transported"]].to_numpy()
X1 = df.loc[8500:8692,["PID","PID3/4","HP1","HP2","HP3","CryoSleep","DB","DC","DC4/5","Side","D1","D2","D3","NormAge","VIP","NormRoomService","NormFoodCourt","NormShoppingMall","NormSpa","NromVRDeck"]].to_numpy()
Y1 = df.loc[8500:8692,["Transported"]].to_numpy()
```

```python
def sigmoid (no):
    e = 2.718281828459045
    return 1/(1+e**(-no))
```

```python
print (sigmoid(0))
```

```python
def sigmoid_inv (no):
    return np.log(no/(1-no))
```

```python
N = 8000
n1 = 40
M = 20
weights_0 = np.random.rand(M,n1) - 0.5
weights_1 = np.random.rand(n1,1) - 0.5
bias_0 = np.random.rand(1,n1) - 0.5
bias_1 = 0
# last column is coeff b/w ans and first layer
# In r th row, cell with column i is the weight from i th input to r th neuron in first layer
# Last row for bias
```

```python
alpha = 0.0001
```

```python
a = 0

while a > -1:
      
    # data -> 1st Layer -> Final
    
    answer_1 = sigmoid(np.dot(X,weights_0) + bias_0)
    answer_2 = sigmoid(np.dot(answer_1,weights_1) + bias_1)

    # Cost Function
    
    Cost = np.sum((f(answer_2) - Y) ** 2)
    if a%10 == 0:
        print(Cost)
        print(np.average(abs((f(answer_2) - Y)*100)))
        answer1 = sigmoid(np.dot(X1,weights_0) + bias_0)
        answer2 = sigmoid(np.dot(answer1,weights_1) + bias_1)
        print(np.average(abs((f(answer2) - Y1)*100)))
        print("\n")
#         print(answer_2)

    # Safety
    if a == 0:
        cost_prev = Cost
    cost_curr = Cost
    if (cost_curr>cost_prev):
        alpha = alpha / 2
        cost_prev = cost_curr
        continue
    cost_prev = cost_curr
    
    # Optimise
    
    if a % 200 == 0:
        alpha = alpha * 1.5
    
    # Update
    
    weights_1 = weights_1 - alpha * np.dot(np.transpose(answer_1),2 * (answer_2 - Y) * answer_2 * (1 - answer_2))
    bias_1 = bias_1 - alpha * np.dot(np.transpose(2 * (answer_2 - Y)),answer_2 * (1 - answer_2))
    weights_0 = weights_0 - alpha * np.dot(np.transpose(X),2 * (answer_2 - Y) * answer_2 * (1 - answer_2) * (np.transpose(weights_1) * answer_1) * (1 - answer_1))
    bias_0 = bias_0 - alpha * np.transpose(np.reshape(np.sum(2 * (answer_2 - Y) * answer_2 * (1 - answer_2) * (np.transpose(weights_1) * answer_1 * (1 - answer_1)),axis=0),[n1,1]))
     
    a = a + 1
```
